## Figure and Axes

Matplotlib has two core objects: the `Figur`e and the `Axes`. The `Axes` object is an individual plot, containing an x-axis, a y-axis, labels, etc.; it also contains all of the various methods we might use for plotting. A `Figure` contains one or more `Axes` objects; it also contains methods for saving plots to files (e.g., PNG, SVG), among other similar high-level functionality.

## Colors

```python
fig = plt.figure(figsize=(18, 8))
ax = fig.add_subplot(1, 1, 1)

# tab:red Tableau red color
ax.plot(times, temps, color='tab:red', label='Temperature (surface)')
ax.plot(
    times,
    temps_1000,
    color='tab:red',
    linestyle='--',
    label='Temperature (isobaric level)'
)

ax.set_xlabel('Time')
ax.set_ylabel('Temperature')
ax.set_title('Temperature Forecast', size=18)

ax.grid(True)

ax.legend(loc='upper right')
```

## Subplots

```python
dewpoint = 0.9 * temps
dewpoint_1000 = 0.9 * temps_1000
```

Use the `.add_subplot()` method to add subplots to our figure! This method takes the arguments `(rows, columns, subplot_number)`.

```python
fig = plt.figure(figsize=(18, 8))

ax = fig.add_subplot(1, 2, 1)
ax.plot(times, temps, color='tab:red')

ax2 = fig.add_subplot(1, 2, 2)
ax2.plot(times, dewpoint, color='tab:green')
```

`pyplot.subplots()` with the keyword arguments `nrows` (number of rows) and `ncols` (number of columns). This initializes a new Axes object, called `ax`, with the specified number of rows and columns. This object also contains a 1-D list of subplots, with a size equal to `nrow`s x `ncols`.

```python
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(18, 8))

ax[0].plot(times, temps, color='tab:red')
ax[1].plot(times, dewpoint, color='tab:green')
```

### Titles on subplots

```python
fig, ax = plt.subplots(1, 2, figsize=(18, 8))

ax[0].plot(times, temps, color='tab:red')
ax[0].set_title('Temperatura')
ax[1].plot(times, dewpoint, color='tab:green')
ax[1].set_title('Dewpoint')
```

## Plot boundaries

Using `ax.set_xlim` and `ax.set_ylim` to control the plot boundaries.

```python
fig = plt.figure(figsize=(18, 8))

ax = fig.add_subplot(1, 2, 1)
ax.plot(times, temps, color='tab:red')
ax.set_title('Temperatures')
ax.set_xlim(110, 130)
ax.set_ylim(290, 315)

ax2 = fig.add_subplot(1, 2, 2)
ax2.plot(times, dewpoint, color='tab:green')
ax2.set_title('Dewpoint')
ax2.set_xlim(110, 130)
```

```python
fig = plt.figure(figsize=(18, 8))

ax = fig.add_subplot(1, 2, 1)
ax.plot(times, temps, color='tab:red')
ax.set_title('Temperatures')
ax.set_ylim(260, 320)

ax2 = fig.add_subplot(1, 2, 2,  sharex=ax, sharey=ax)
ax2.plot(times, dewpoint, color='tab:green')
ax2.set_title('Dewpoint')
```

```python
fig, ax = plt.subplots(1, 2, figsize=(18, 8), sharex=True, sharey=True)

# Temperature
ax[0].plot(times, temps, color='tab:red', label='Temperatures (surface)')
ax[0].plot(times, temps_1000, color='tab:red', linestyle='--', marker='o', label='Temperatures (isobaric level')
ax[0].set_xlabel('Time')
ax[0].set_ylabel('Temperature')
ax[0].set_title('Temperature Forecast', size=16)
ax[0].grid(True)
ax[0].legend(loc='lower right')
ax[0].set_xlim(100, 160)
ax[0].set_ylim(260, 310)

# Dewpoint
ax[1].plot(times, dewpoint, color='tab:green', label='Dewpoint (surface)')
ax[1].plot(times, dewpoint_1000, color='tab:green', linestyle='--', marker='o', label='Dewpoint (isobaric level')
ax[1].set_xlabel('Time')
ax[1].set_ylabel('Dewpoint')
ax[1].set_title('Dewpoint Forecast', size=16)
ax[1].grid(True)
ax[1].legend(loc='upper left')
```

## Scatterplot

Creating a scatter plot by using a regular plot and formating the line and markers.

```python
fig = plt.figure(figsize=(18, 8))
ax = fig.add_subplot(1, 1, 1)

ax.plot(temps, temps_1000, linestyle='None', marker='o', markersize=8)

ax.set_xlabel('Temperature (surface)')
ax.set_ylabel('Temperature (1000 hPa)')
ax.set_title('Temperature Cross Plot')
ax.grid(True)
```

Creatting a scatter plot with the method `scatter`

```python
fig = plt.figure(figsize=(18, 8))
ax = fig.add_subplot(1, 1, 1)

ax.set_xlabel('Temperature (surface)')
ax.set_ylabel('Temperature (1000 hPa)')
ax.set_title('Temperature Cross Plot')
ax.scatter(temps, temps_1000)
ax.grid(True)
```

```python
fig = plt.figure(figsize=(18, 8))
ax = fig.add_subplot(1, 1, 1)

ax.plot([285, 320], [285, 320], color='black', linestyle='--')
s = ax.scatter(temps, temps_1000, c=(temps - temps_1000), cmap='bwr', vmin=-5, vmax=5)
fig.colorbar(s)

ax.set_xlabel('Temperature (surface)')
ax.set_ylabel('Temperature (1000 hPa)')
ax.set_title('Temperature Cross Plot')
ax.grid(True)
```

## Displaying images

```python
x = y = np.arange(-3.0, 3.0, 0.025)
X, Y = np.meshgrid(x, y)
Z1 = np.exp(-(X**2) - Y**2)
Z2 = np.exp(-((X - 1) ** 2) - (Y - 1) ** 2)
Z = (Z1 - Z2) * 2
```

```python
fig, ax = plt.subplots(figsize=(18, 8))
im = ax.imshow(Z, interpolation='bilinear', cmap='RdYlGn', origin='lower', extent=[-6, 6, -3, 3])
```

```python
fig, ax = plt.subplots(figsize=(18, 8))
ax.contour(X, Y, Z)
```

```python
fig, ax = plt.subplots(figsize=(18, 8))
c = ax.contour(X, Y, Z, levels=np.arange(-2, 2, 0.25))
ax.clabel(c)
```

```python
fig, ax = plt.subplots(figsize=(18, 8))
c = ax.contourf(X, Y, Z)
```

```python
fig, ax = plt.subplots(figsize=(18, 8))
im = ax.imshow(
    Z, interpolation='bilinear', cmap='PiYG', origin='lower', extent=[-3, 3, -3, 3]
)
c = ax.contour(X, Y, Z, levels=np.arange(-2, 2, 0.5), colors='black')
ax.clabel(c)
```
