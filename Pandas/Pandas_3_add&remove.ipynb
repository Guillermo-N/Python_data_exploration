# Add and remove operation with Pandas

```python
import pandas as pd
```

```python
df = pd.read_csv('../People_data.csv')
```

```python
pd.set_option('display.max_columns', None)
```

```python
df
```

```python
# Adding a new column to the dataframe
df['Fullname'] = df['Name'] + ' ' + df['Surname']
df
```

```python
# Dropping the addes column from the dataframe
df.drop(columns=['Fullname'], inplace=True)
df
```

```python
df[['Surname_1', 'Surname_2']] = df['Surname'].str.split(' ', expand=True)
df
```

## Adding rows

```python
df = df.append({'Name': 'Tony'}, ignore_index=True)
```

```python
df_2 = pd.DataFrame({
    'Name': ['Javi', 'Antonio'],
    'Surname': ['Tudor Nacil', 'Silva Losas'],
    'Email': ['javi@gamil.com', 'asl@gamil.com']
})
```

```python
df = df.append(df_2, ignore_index=True)
df
```

```python
df.drop(index=6, inplace=True)
df
```
